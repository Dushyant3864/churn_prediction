```python
# Remove newly joined customers
df = df[df["Customer Status"] != "Joined"]

# Create binary target
df["Churn Status"] = df["Customer Status"].map({
    "Stayed": "No",
    "Churned": "Yes"
})
```

**Excluding Newly Joined Customers from Churn Modeling**

The dataset contained three customer status categories: Stayed (4720), Churned (1869), and Joined (454). The Joined customers were excluded before creating the final target variable because they represent newly acquired customers who may not have had sufficient time or opportunity to churn.

Including these customers as non-churned could introduce label bias, as their current status does not necessarily mean they are loyal customers; it only means they have recently joined and have not yet shown churn behaviour. Since they form only 454 out of 7043 customers, approximately 6.4% of the dataset, removing them does not significantly reduce the dataset size. After excluding them, the model is trained only on customers with clear churn outcomes: Stayed and Churned.

```python
#Encoding the Target Variable into Binary Format
df["Churn_Binary"] = df["Churn Status"].map({"No": 0,"Yes": 1})
```

The Churn Status column was converted into a binary numerical format by mapping "No" to 0 and "Yes" to 1 using the map() function.

Since most classification models require the target variable to be represented numerically, this transformation enables the model to learn and predict customer churn effectively.

```python
df["Avg Monthly Long Distance Charges"].corr(df["Churn_Binary"])
```

```python
from scipy.stats import ttest_ind

no = df[df["Churn_Binary"] == 0]["Avg Monthly Long Distance Charges"].dropna()
yes = df[df["Churn_Binary"] == 1]["Avg Monthly Long Distance Charges"].dropna()

t_stat, p_value = ttest_ind(no, yes)

print(p_value)
```

The correlation between Avg Monthly Long Distance Charges and the binary target variable Churn_Binary was calculated using the corr() function to measure the strength and direction of their linear relationship. Additionally, an independent two-sample t-test (ttest_ind) was performed to determine whether the average monthly long-distance charges differed significantly between customers who churned and those who did not.

 **Observation:**

- The correlation coefficient obtained was −0.00047, which is extremely close to 0, indicating no meaningful linear relationship between average monthly long-distance charges and customer churn.
- The independent two-sample t-test produced a p-value of 0.9713, which is substantially greater than the significance level of 0.05.
- Therefore, the null hypothesis cannot be rejected, indicating that there is no statistically significant difference in the average monthly long-distance charges between churned and non-churned customers.
- Based on both the correlation analysis and hypothesis testing, Avg Monthly Long Distance Charges has no meaningful predictive relationship with customer churn and was therefore removed from the dataset.
